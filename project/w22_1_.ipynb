```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python

def V(x):
    return -math.exp(-x**2)-1.2 * math.exp(-(x-2)**2)

def potencjal(x):
    return 2*math.exp(-x**2)* x + 2.4 * math.exp(-(x - 2)**2) * (x - 2)
def polozenie(x,v,m,time):
    x_t = [x]
    v_t = [v]
    delta = time[1]-time[0]
    for i in range(len(time)-1):
        v_t.append(
            v_t[i] - 1/m * potencjal(x_t[i]) * delta
        )
        x_t.append(
            x_t[i] + v_t[i]*delta
        )
    return [x_t, v_t]

def polozenieOpory(x,v,m,time,alfa):
    x_t = [x]
    v_t = [v]
    delta = time[1]-time[0]
    for i in range(len(time)-1):
        v_t.append(
            v_t[i] - 1/m * potencjal(x_t[i]) * delta - alfa * v_t[i] * delta
        )
        x_t.append(
            x_t[i] + v_t[i]*delta
        )
    return [x_t, v_t]
    
```

```python

m = 1 #kg
v = 0
x = 2.8 #m
n = 0
delta_t_1 = 0.01
delta_t_2 = 0.001
t_1 = 100
t_2 = 1000
time_11 = np.arange(0,t_1,delta_t_1)
time_12 = np.arange(0,t_1,delta_t_2)
time_21 = np.arange(0,t_2,delta_t_1)
time_22 = np.arange(0,t_2,delta_t_2)
```

```python
t_1_dt_1 = np.array(polozenie(x,v,m,time_11))

plt.subplot(5,1,1)
plt.plot(time_11,t_1_dt_1[0])
plt.xlim(0,100)
plt.title("x,v, $E_k$, dla t = (0,100) oraz $\Delta$ t = 0.01")
plt.ylabel("x")

plt.subplot(5,1,2)
plt.plot(time_11,t_1_dt_1[1])
plt.xlim(0,100)
plt.ylabel("v")

plt.subplot(5,1,3)
plt.plot(time_11,t_1_dt_1[1]**2 * m * 1/2)
plt.xlim(0,100)
plt.ylabel("$E_k$")

potencjalna = []
for i in t_1_dt_1[0]:
    potencjalna.append(V(i))

plt.subplot(5,1,4)
plt.plot(time_11,potencjalna)
plt.xlim(0,100)
plt.ylabel("$V$")

plt.subplot(5,1,5)
plt.plot(time_11,np.array(potencjalna) + t_1_dt_1[1]**2 * m * 1/2)
plt.xlim(0,100)
plt.ylabel("$V + E_k$")
```
