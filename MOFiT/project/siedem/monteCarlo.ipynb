```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def ro(x):
    return math.pow(np.pi,-0.5)*math.exp(-math.pow(x,2))

def MomentRozkładu(X,l,n):
    return np.sum(np.power(X[:l],n))/l
```

```python
N = 10000000
X_W = np.zeros(N)
x_w = 0
x_n = 0
l = 0
data_4 = np.random.rand(N) * 0.5 - 0.25
data_1 = np.random.rand(N)


for i in range(N):
    x_n = x_w + data_4[i]
    if data_1[i] < (ro(x_n)/ro(x_w)):
        x_w = x_n
        l += 1
    X_W[i] = x_w

```

```python
points = 100
I_n = np.zeros((4,points))
j = 0
for i in np.logspace(1,7,points):
    i = int(i)
    I_n[0,j] = MomentRozkładu(X_W,i,1)
    I_n[1,j] = MomentRozkładu(X_W,i,2)
    I_n[2,j] = MomentRozkładu(X_W,i,3)
    I_n[3,j] = MomentRozkładu(X_W,i,4)
    j +=1
```

```python
plt.figure(figsize=(10,10))

plt.subplot(4,1,1)
plt.plot(I_n[0])
plt.plot([0]*points)
plt.ylabel('n = 1')

plt.subplot(4,1,2)
plt.plot(I_n[1])
plt.plot([0.5]*points)
plt.ylabel('n = 2')


plt.subplot(4,1,3)
plt.plot(I_n[2])
plt.plot([0]*points)
plt.ylabel('n = 3')


plt.subplot(4,1,4)
plt.plot(I_n[3])
plt.plot([0.75]*points)
plt.ylabel('n = 4')
plt.xlabel('numer próbki, rozkład logarytmiczny $[0,10^7]$')

```

```python
def phi(x,y):
    return math.pow(math.pow(np.pi,-0.5) * math.exp(-0.5*(math.pow(x,2)+math.pow(y,2))),2)

def energiaPot(X,Y,l):
    return np.sum((np.power(X[:l],2) + np.power(Y[:l],2))/2)/l
```

```python
N = 10000000
X_W_2D = np.zeros(N)
Y_W_2D = np.zeros(N)
x_w = 0
y_w = 0
x_n = 0
y_n = 0
l = 0
data_4_x = np.random.rand(N) * 0.5 - 0.25
data_4_y = np.random.rand(N) * 0.5 - 0.25
data_1 = np.random.rand(N)


for i in range(N):
    x_n = x_w + data_4_x[i]
    y_n = y_w + data_4_y[i]
    if data_1[i] < (phi(x_n,y_n)/phi(x_w,y_w)):
        x_w = x_n
        y_w = y_n
        l += 1
    X_W_2D[i] = x_w
    Y_W_2D[i] = y_w
```

```python
points = 100
En = np.zeros((points))
j = 0
for i in np.logspace(1,7,points):
    i = int(i)
    En[j] = energiaPot(X_W_2D,Y_W_2D,i)
    j += 1
```

```python
plt.figure(figsize=(10,10))
plt.plot(En)
plt.plot([0.5]*points)
plt.title('wartość energi potencjalnej')
plt.ylabel('En')
plt.xlabel('numer próbki, rozkład logarytmiczny $[0,10^7]$')
```
